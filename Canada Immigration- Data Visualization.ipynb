```python
DSN=canada.loc[['Denmark','Norway','Sweden'],:]
DSN
```

```python
total=sum(DSN['Total'])
category_proportions=[(float(value)/total) for value in DSN['Total']]
category_proportions # total will be proportioned in 3 values
```

```python
# Define overall size of waffle chart
width=40
height=10
total_tiles=width*height
print('Total number of tiles is ', total_tiles)
```

```python
# Step 3: give values according proportion
tiles_per_category=[round(proportion*total_tiles) for proportion in category_proportions]
tiles_per_category
```

```python
# initialize the waffle chart as an empty matrix
waffle_chart=np.zeros((height,width))

# define indices to loop through waffle chart
category_index=0
tile_index=0

# populate the waffle chart
for col in range(width):
  for row in range(height):
    tile_index=tile_index+1
    #if number of tiles populated for the current category is equal to its corresponding allocated tiles.
    if tile_index> sum(tiles_per_category[0:category_index]):
        # processed to the next category
        category_index=category_index+1
      #set the class value to an integer, which increse with class
    waffle_chart[row,col]=category_index

print ('Waffle chart populated!')
```

```python
 # instantiate a new figure object
fig=plt.figure()

# Use matshow to display the waffle chart
colormap=plt.cm.coolwarm
plt.matshow(waffle_chart,cmap=colormap)
plt.colorbar()
```

```python
# intiate a new figure object
fig=plt.figure()
# Use matshow to display the waffle chart
colormap=plt.cm.coolwarm
plt.matshow(waffle_chart,cmap=colormap)
plt.colorbar()

# get the axis
ax=plt.gca()
#set minor ticks
ax.set_xticks(np.arange(-.5,(width),1),minor=True)
ax.set_yticks(np.arange(-.5,(height),1),minor=True)
#add gridlines based on minor ticks
ax.grid(which='minor',color='w',linestyle='-',linewidth=2)
plt.xticks([])
plt.yticks([])
plt.show()
```

# plot regplot for Total Immigration to Canada from 1980-2013


```python
# we can use the sum() method to get the total population per year
total = pd.DataFrame(canada[years].sum(axis=0))
# change the years to type float (useful for regression later on)
total.index = map(float,total.index)
# reset the index to put in back in as a column in the df_tot dataframe
total.reset_index(inplace = True)
 
# rename columns
total.columns = ['year', 'total']
total
```

```python
import seaborn as sns
ax = sns.regplot(x='year', y='total', data=total,color='green')
```

```python
plt.figure(figsize=(10, 8))
ax = sns.regplot(x='year', y='total', data=total, color='green', marker='+', scatter_kws={'s': 200})
 
ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-labels
ax.set_title('Total Immigration to Canada from 1980 - 2013') # add title
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.regplot(x='Year', y='Total', data=canada_DNS_regplot, color='green')
 
ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-labels
ax.set_title('Total Immigration to Canada from 1980 - 2013') # add title
```
